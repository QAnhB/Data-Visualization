```python
# advance!!!
# Note: You need to install Matplotlib, scipy package!
# Lorentz attractor

import numpy as np
import matplotlib.pyplot as plt


def lorenz(xyz, *, s=10, r=28, b=2.667):
    """
    Parameters
    ----------
    xyz : array-like, shape (3,)
       Point of interest in three dimensional space.
    s, r, b : float
       Parameters defining the Lorenz attractor.

    Returns
    -------
    xyz_dot : array, shape (3,)
       Values of the Lorenz attractor's partial derivatives at *xyz*.
    """
    x, y, z = xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return np.array([x_dot, y_dot, z_dot])


dt = 0.01
num_steps = 10000

xyzs = np.empty((num_steps + 1, 3))  # Need one more for the initial values
xyzs[0] = (0., 1., 1.05)  # Set initial values
# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point
for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt

# Plot
ax = plt.figure().add_subplot(projection='3d')

ax.plot(*xyzs.T, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

```python
# advance!!!
# Note: You need to install Matplotlib, scipy package!

import matplotlib.pyplot as plt
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from scipy.special import sph_harm
%matplotlib inline

def spherical_coordinate(phi, theta):   # this function can be merged into spherical harmonic function!
    # The Cartesian coordinates of the unit sphere
    x = np.sin(phi) * np.cos(theta)
    y = np.sin(phi) * np.sin(theta)
    z = np.cos(phi)

    return x,y,z


def spherical_harmonics(m,l, theta,phi):
    ###===== Calculate the spherical harmonic Y(l,m) and normalize to [0,1].
    fcolors = sph_harm(m, l, theta, phi).real    # real part 
    fmax, fmin = fcolors.max(), fcolors.min()
    fcolors = (fcolors - fmin) / (fmax - fmin)
        
    return fcolors


##=============== Declare variables and arguments
phi = np.linspace(0, np.pi, 100)        # linear space
theta = np.linspace(0, 2*np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

m, l = 2, 3   # Note: l>m;  m: order ; l >0: Degree of spherical harmonics.
#m, l = 0,3

###=============== Set up a Cartesian coordinate and calculate the spherical harmonic Y(l,m)
x,y,z = spherical_coordinate(phi, theta)
fcolors = spherical_harmonics(m,l,theta,phi)

###================ Plot (Visualization lecture)
fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z , facecolors=cm.seismic(fcolors) ) #, rstride=1, cstride=1 )  # Set the aspect ratio to 1 so our sphere looks spherical

##===== Turn off the axis planes
#ax.set_axis_off()
ax.tick_params(axis='both', which='both', labelsize=22, pad=22)
plt.show()
```

- A `class` can contain variables, functions, loops, etc. A Python class is created with a key word `class`.
- In a `class`, the `__init__` function is used to assign values, properties, operation. This function is called automatically when `class` is used.

- `self` parameter is used to acess variable in the `class`. You can change this what you want example `my_first_var`, etc.
- When a `class` is called, it is an `object`.

```python

class Person_Class:
    def __init__(self, fname, lname, age):
        '''
        This init function can be  used to initiate all variables.
        The function after just need to have self parameter.
        '''
         
        self.fname = fname
        self.lname = lname
        self.age = age
        
        
    def lastname(self):
        print("Hello my last name is " + self.lname)
```

```python
## When we call class, it is object, this will remind you about C language.
Object = Person_Class("USTH", "Space", 12)


## Call a function in class
Object.lastname()


print(Object.fname)
print(Object.age)


# Modify object properties
Object.age = 100
print('Modified param', Object.age)


# Delete object
del Object.age
#print('Deleted param', Object.age)
```

### Colormap and contour plots

- There are several similar function for color plot: `pcolor`, `pcolormesh`, `imshow`, and `pcolorfast`.
- `mgrid`  
- `meshgrid`

```python
# make these smaller to increase the resolution
dx, dy = 0.15, 0.15

# generate 2 2d grids for the x & y bounds
Y, X = np.mgrid[-3:3+dy:dy, -3:3+dx:dx]
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# x and y are bounds, so z should be the value *inside* those bounds.
# Therefore, remove the last value from the z array.
Z = Z[:-1, :-1]
z_min, z_max = -abs(Z).max(), abs(Z).max()


fig, axs = plt.subplots(2, 2, figsize=(14,6))
#fig.subplots_adjust(left=0.15, right=1.5, bottom=0.1, top=0.9)


ax = axs[0, 0]
c = ax.pcolor(X, Y, Z, cmap='RdBu', vmin=z_min, vmax=z_max)
ax.set_title('pcolor')
fig.colorbar(c, ax=ax)

ax = axs[0, 1]
c = ax.pcolormesh(X, Y, Z, cmap='RdBu', vmin=z_min, vmax=z_max)
ax.set_title('pcolormesh')
fig.colorbar(c, ax=ax)


ax = axs[1, 0]
c = ax.imshow(Z, cmap='RdBu', vmin=z_min, vmax=z_max,
              extent=[X.min(), X.max(), Y.min(), Y.max()],
              interpolation='nearest', origin='lower', aspect='auto')
ax.set_title('image (nearest, aspect="auto")')
fig.colorbar(c, ax=ax)


ax = axs[1, 1]
c = ax.pcolorfast(X, Y, Z, cmap='RdBu', vmin=z_min, vmax=z_max)
ax.set_title('pcolorfast')
fig.colorbar(c, ax=ax)


fig.tight_layout()
plt.show()
```

### 3D plot
In matplotlib, we need to create `Axes3D` class for a 3D plot. This can be added to a matplotlib figure by passing a `projection='3d'` keyword argument to the `add_axes` or `add_subplot` methods.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
```

```python
# plot a 3D surface like in the example mplot3d/surface3d_demo
N=100
vmin=-1.   # chan this value to observe the shape of the object
vmax=1.

X = np.linspace(vmin, vmax, N)   
Y = np.linspace(vmin, vmax, N)

X, Y = np.meshgrid(X, Y)
R = np.sqrt( X**2 + Y**2)
Z = np.sin(R)


# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=(12,12))
fig.subplots_adjust(left=0.15, right=1.5, bottom=0.1, top=0.9)

#===============
#  First subplot
#===============
# set up the axes for the first plot
ax = fig.add_subplot(2, 2, 1, projection='3d')

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

cset = ax.contour(X, Y, Z, zdir='z', offset=0.005, cmap=matplotlib.cm.coolwarm)   # you need to change the offset if you change the data dimension
cset = ax.contour(X, Y, Z, zdir='x', offset=-1, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=1.5, cmap=matplotlib.cm.coolwarm)

#ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5 , aspect=10)
ax.set_title('3D surface')


#===============
# Second subplot
#===============
# set up the axes for the second plot
ax = fig.add_subplot(2, 2, 2, projection='3d')

# plot a 3D wireframe like in the example mplot3d/wire3d_demo
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
ax.set_title('3D wire frame')


#===============
# Third subplot
#===============
# set up the axes for the third plot
ax = fig.add_subplot(2, 2, 3, projection='3d')

cset = ax.contourf(X, Y, Z, zdir='z', cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, levels=14, linewidths=0.5, colors='b')

ax.set_title('3D contour')


#===============
# Third subplot
#===============
ax = fig.add_subplot(2, 2, 4, projection='3d')
n_contour = 50
ax.contour3D(X, Y, Z, n_contour, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')

plt.show()
```
